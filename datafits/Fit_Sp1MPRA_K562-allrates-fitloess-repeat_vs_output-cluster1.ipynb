```python
def plot_best(best,args_fitnessfunction=None):
    #best is the parameter set to be plotted
    afcat_idxs=args_fitnessfunction["afcat_idxs"]
    ncol=len(afcat_idxs) #categories of affinities fitted
    error,arv=newerror(best,**args_fitnessfunction)
    #display(out)
    colors=["blue","cyan","lightgreen","orange","red","purple"]
    fig,axes=plt.subplots(1,ncol,figsize=(15,4),sharey=True)
    for a,idxs in enumerate(afcat_idxs):
        subdf=arv[idxs]
        xvals=np.arange(1,7)
        yvals=subdf[:,0]
        ymodelvals=subdf[:,1]
        fivals=subdf[:,2]
        ax=axes[a]
        color=colors[a]
        ax.scatter(xvals, yvals,c=color)
        ax.plot(xvals,ymodelvals,color=color)
        ax2=ax.twinx()
        ax2.plot(xvals,fivals)
        ax.set_xlabel("nrepeats")
        ax.set_ylabel("log fold change")
        ax2.set_ylabel("fi")
    plt.tight_layout()
    plt.show()
```

```python
def newerror(individual,S_max=1,coop=False,rates_affected=[],logfcfunction=None, evaluate_parset=None,fifunctions={},
                   data=None,  means=[],afcat_idxs=[],
                   idx1=0, idx2=0, idx3=0, plot=False,plotparset=False,nrow=1,ncol=1,parsetnames=[],outerror="inverse",individualerror=False,wlist=None,relrange=False):
    #afcat_idxs:list of lists, each list hast the idxs of the rows that correspond to the same affinity
    ar=np.asarray(individual)
    global_occupancy_pars=ar[0:idx1] #note that now the TF concentration is for each condition, because conditions correspond to celltypes
    print("global_occupancy_pars", global_occupancy_pars)
    
    globalcyclepars=ar[idx1:idx2]
    globalTFcyclepars=ar[idx2:idx3]
    TFcyclepars=globalTFcyclepars
    occupancy_pars=global_occupancy_pars
    values_data_model=[]
    for i in range(len(data)):
        afcat, nrepeats, site1, site2, site3, site4, site5, site6, value= data.iloc[i]
        if nrepeats==1:
            fifunction=fifunctions[1]
            color="navy"
            input_=np.array([site1])
        elif nrepeats==2:
            fifunction=fifunctions[2]
            input_=np.array([site1,site2])
            color="blue"
        elif nrepeats==3:
            fifunction=fifunctions[3]
            input_=np.array([site1,site2,site3])
            color="cyan"
        elif nrepeats==4:
            fifunction=fifunctions[4]
            input_=np.array([site1,site2,site3,site4])
            color="lightgreen"
        elif nrepeats==5:
            fifunction=fifunctions[5]
            input_=np.array([site1,site2,site3,site4,site5])
            color="green"
        elif nrepeats==6:
            fifunction=fifunctions[6]
            input_=np.array([site1,site2,site3,site4,site5,site6])
            color="olive"
        else:
            raise ValueError("Wrong nrepeats")
        fi=fifunction(pars=occupancy_pars,S_i=input_,S_max=S_max,coop=coop,output="avbinding") #returns average number of bound molecules on DNA
        
        #fi=np.sum(fi)
        K_max, lambda_K=np.exp(global_occupancy_pars[0:2])
        print(afcat, "input", input_, "fi",fi, K_max*np.exp((input_-S_max)/lambda_K))
        output=logfcfunction(globalpars=globalcyclepars,TFpars=TFcyclepars,avbinding=fi,rates_affected=rates_affected)
        values_data_model.append([value,output,fi])
    arv=np.asarray(values_data_model)
    E_list=[]
    for i,idxs in enumerate(afcat_idxs):
        w=wlist[i]
        affinity_points=arv[idxs,:2]
        #print(affinity_points)
        if relrange:
            range_=np.ptp(affinity_points[:,0])
        else:
            range_=1
        error=np.abs(np.diff(affinity_points,axis=1))/range_
        E_list.append(w*error)
    #print("E_list",E_list)
    total_error=np.sum(E_list)
    return [total_error,arv]
```
