```r
library(ggplot2)
library(reshape)
```

```r
#plasmid conc-GFP ouptut
inputconc=c(0.125,0.25,0.5,0.75)
SP1_1=c(1.68559,2.84942,3.74315,2.76530)
SP1_2=c(0.71226,1.95223,2.62698,1.08696)

df<-data.frame(inputconc,SP1_1,SP1_2)
df
```

```r
melted<-melt(df,id="inputconc")
melted
```

```r
ggplot(df, aes(x = inputconc)) +
  geom_point(aes(y = SP1_1), color = "blue") +
  geom_point(aes(y = SP1_2), color = "red") +
  geom_smooth(aes(y = (SP1_1 + SP1_2) / 2), method = "loess", color = "green", span=0.8) +
  labs(x = "Input Concentration", y = "Value", title = "Scatterplot with Smoothed Mean") +
  theme_classic()
```

```r
#plasmid conc - inputZF


SP1_1_i=c(0.02466,0.05381,0.34476,0.46896) #normalised expression, not fold change. 
SP1_1_i=SP1_1_i/SP1_1_i[1]
SP1_2_i=c(0.80526,0.37440,6.58052,7.49619)
SP1_2_i=SP1_2_i/SP1_2_i[1]
```

```r
df<-data.frame(inputconc,SP1_1_i,SP1_2_i)
df
```

```r
melted<-melt(df,id="inputconc")
melted
```

```r
ggplot(df, aes(x = inputconc)) +
  geom_point(aes(y = SP1_1_i), color = "blue") +
  geom_point(aes(y = SP1_2_i), color = "red") +
  geom_smooth(aes(y = (SP1_1_i + SP1_2_i) / 2), method = "loess", color = "green") +
  labs(x = "Input Concentration", y = "Value", title = "Scatterplot with Smoothed Mean") +
  theme_classic()
```
