```python
#will convert parameters above into notation in paper
a_x,b_x,a_p,b_p,alpha,beta,gamma,omega,k_o,k_c=sympy.symbols("a_x,b_x,a_p,b_p,alpha,beta,gamma,omega,k_o,k_c")
ax_0_0,ax_0_1,bx_x_0,bx_x_1,aP_0_0,aP_0_1,bP_P_0,bP_P_1,aP_x_0,aP_x_1,bP_xP_0,bP_xP_1,ax_P_0,ax_P_1,bx_xP_0,bx_xP_1,kopen0,kclose0,kopenx,kclosex,kopenP,kcloseP,kopenxP,kclosexP=sympy.symbols("ax_0_0,ax_0_1,bx_x_0,bx_x_1,aP_0_0,aP_0_1,bP_P_0,bP_P_1,aP_x_0,aP_x_1,bP_xP_0,bP_xP_1,ax_P_0,ax_P_1,bx_xP_0,bx_xP_1,kopen0,kclose0,kopenx,kclosex,kopenP,kcloseP,kopenxP,kclosexP")
q3,q4,q7,q8=sympy.symbols("q3,q4,q7,q8")

# (note that omega and gamma in the paper are 1/omega and 1/gamma here)

replace_dict={ax_0_0:a_x,bx_x_0:b_x,aP_0_0:a_p,bP_P_0:b_p,
             ax_0_1:a_x,bx_x_1:b_x,aP_0_1:a_p,bP_P_1:b_p*beta,
             ax_P_0:a_x,bx_xP_0:b_x*omega,aP_x_0:a_p,bP_xP_0:b_p*omega,
             ax_P_1:a_x,bx_xP_1:b_x*gamma*omega,aP_x_1:a_p,bP_xP_1:b_p*beta*gamma*omega,
             kopen0:k_o,kclose0:k_c,kopenx:alpha*k_o,kclosex:k_c,
              kopenP:k_o,kcloseP:k_c*beta,kopenxP:alpha*k_o,kclosexP:k_c*beta*gamma}
```

```python
def write_file_tocheck(idxlist_num=[],variables=[],folder=None,file=None):
    
    #make a list where each item k is the coefficient of degree k
    #qs={3:q3,4:q3*q4,7:q3*q4*q7,8:q3*q4*q7*q8} #q_i>=1 #since q8>q7>q4>q3, I assume q4=q4*q3, q4>1, q7=q3*q4*q7, q7>1, and same for q8. So when checking with mathematica I can use the constraint with 1 rather than among themselves
    #for i>3, this would correspond to the epsilon_i in the paper.
    print("calculating num and den")
    sys.stdout.flush()
    
    degrees_num=[]
    for k in range(5):
        coeffs=0
        for i in idxlist_num: #rhos in the numerator. rho3:2,rho4:3,rho7:6,rho8:7 
            rho=obj.coeffs_rhos[i]
            if k<=len(rho)-1:
                coeffs+=rho[k] 
            else:
                print("Missing degree %d for rho %d"%(k,i+1))
        degrees_num.append(coeffs)

    degrees_den=[]
    for k in range(5):
        coeffs=0
        for i in range(8): #all rhos
            rho=obj.coeffs_rhos[i]
            if k<=len(rho)-1:
                coeffs+=rho[k]
            else:
                print("Missing degree %d for rho %d"%(k,i+1))
        degrees_den.append(coeffs)
        
    print("replacing notation in num and den")
    sys.stdout.flush()
    
    #replace parameters by notation defined in replace_dict, accounting for the fact that many are the same
    new_degrees_num=[]
    for item in degrees_num:
        if item==0:
            newitem=item
        else:
            newitem=item.subs(replace_dict)
        new_degrees_num.append(newitem)

    new_degrees_den=[]
    for item in degrees_den:
        newitem=item.subs(replace_dict)
        new_degrees_den.append(newitem)
        
    #now prepare each combination bi aj - ai bj. We need to check that each combination is >=0 for some with inequality
    print("preparing i,j combinations")
    sys.stdout.flush()
    
    combis_ij=[]
    for i in range(5):
        a_i=new_degrees_num[i]
        b_i=new_degrees_den[i]
        #print(len(a_i),len(b_i))

        for j in range(i+1,5):
            print(i,j)
            a_j=new_degrees_num[j]
            b_j=new_degrees_den[j]
            #print(len(a_j),len(b_j))
            plus=sympy.expand(b_i*a_j)
            minus=sympy.expand(a_i*b_j)
            if minus!=0:
                dif=plus-minus
                #dif=sympy.collect(plus-minus,q3)/q3
                print(i,j, "%d monomials"%len(sympy.Add.make_args(dif)))
                combis_ij.append(dif)
            else:
                print("no negative contribution:", minus)
    
    if not os.path.isdir(folder):
        os.mkdir(folder)
    
    filename=os.path.join(folder,file)
    print("writing to file", filename)
    sys.stdout.flush()
    write_expressions_formathematica.write_expressions_formathematica(filename,combis_ij,variables,printi=100) #printi=100 will not print anything

    
```

With this code, we can write down the expressions to check the monotonicity conditions explained in the supplementary of the paper for each of the cases (e.g. P3'+P4'+P7'+P8', or only P8' etc), each of which is done in a different folder (below).

In each folder, there is a mathematica script (e.g. testpositive.wl), where the number of lines has been adjusted to the number of lines in the .in file, and the assumptions adjusted for activation/repression. 

Then execute wolframscript -file testpositive.wl.

Then from within a given folder, do in bash: egrep "True" results/*.out | wc -l --> this will give a number that is the same as the number of lines in the file filename. Which means that mathematica could check that each expression is >0 for activation, of <0 for repression.

Can also double check doing egrep -v "True" results/\*.out | egrep -v "#" -->
In some cases, there are a few lines that don't finish in the first round due to taking too long, see the "\*manual.nb" notebooks where more specific adjustements were done to finally be able to doubleckeck those lines as well.
