```python
import numpy as np
import matplotlib.pyplot as plt
import sys,os
sys.path.append("../cflinux/")
sys.path.append("../cfmac")
import PolAB_A_allpars
```

```python
#Node correspondance:

nodedict={1:"a0",2:"i0",3:"n0",4:"aB",5:"iB",6:"nB",7:"aA",8:"iA",9:"nA"}
```

```python
basal=np.array([10,1,100,0.1])
alphas=np.array([5,2,0.1,5])
binding=np.array([1,1,1,1,1,1])
unbinding=np.array([1,1,1,1,1,1])
parset=np.ones(24)
parset[0:4]=basal
parset[4:8]=alphas*basal
parset[12:24:2]=binding
parset[13:24:2]=unbinding
B=np.array([0])
Arange=np.logspace(-2,6,1000)
log10Arange=np.log10(Arange)
out=[PolAB_A_allpars.interface_rhos_GRF_PolAB_A_A(parset,B,A) for A in Arange]
outar=np.vstack(out)

ss=outar/np.sum(outar,axis=1)[:,None]
ktan0=parset[1]
ktanA=parset[5]
fig,axes=plt.subplots(1,2,figsize=(10,4))
ax=axes[0]
ax.plot(log10Arange,ss[:,0]*ktan0+ss[:,6]*ktanA)
ax.set_title("GRF")
ax=axes[1]
ax.plot(log10Arange,ss[:,0]+ss[:,6])
ax.set_title("ssn0+ssnA")
plt.show()
fig,axes=plt.subplots(2,3,figsize=(10,5))
n=0
for i in [2,1,0,8,7,6]:
    ax=axes[n//3][n%3]
    ax.plot(log10Arange,ss[:,i],label=nodedict[i+1])
    ax.legend()
    n+=1
plt.tight_layout()
plt.show()
```

Notice that a0 is nonmonotonic but still the result is monotonic!!

Now check a nonmonotonic case:

```python
basal=np.array([10,1,100,0.1])
alphas=np.array([1,0.1,0.1,10])
binding=np.array([1,1,1,1,1,1])
unbinding=np.array([1,1,1,1,1,1])
parset=np.ones(24)
parset[0:4]=basal
parset[4:8]=alphas*basal
parset[12:24:2]=binding
parset[13:24:2]=unbinding
B=np.array([0])
Arange=np.logspace(-2,4,100)
log10Arange=np.log10(Arange)
out=[PolAB_A_allpars.interface_rhos_GRF_PolAB_A_A(parset,B,A) for A in Arange]
outar=np.vstack(out)

ss=outar/np.sum(outar,axis=1)[:,None]
ktan0=parset[1]
ktanA=parset[5]
fig,axes=plt.subplots(1,2,figsize=(10,4))
ax=axes[0]
ax.plot(log10Arange,ss[:,0]*ktan0+ss[:,6]*ktanA)
ax.set_title("GRF")
ax=axes[1]
ax.plot(log10Arange,ss[:,0]+ss[:,6])
ax.set_title("ssn0+ssnA")
plt.show()
plt.plot(log10Arange,np.sum(ss[:,0:3],axis=1),label='unbound')
plt.plot(log10Arange,np.sum(ss[:,6:],axis=1),label='bound')
plt.legend()
plt.title("unbound vs bound states")
fig,axes=plt.subplots(2,3,figsize=(10,5))
n=0
for i in [2,1,0,8,7,6]:
    ax=axes[n//3][n%3]
    ax.plot(log10Arange,ss[:,i],label=nodedict[i+1])
    ax.legend()
    n+=1
plt.tight_layout()
plt.show()
```

In this case aA also increases monotonically and a0 nonmonotonically but still the output is nonmonotonic.
