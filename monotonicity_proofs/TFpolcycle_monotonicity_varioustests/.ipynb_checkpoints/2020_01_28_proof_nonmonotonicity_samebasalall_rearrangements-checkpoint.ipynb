```python
import numpy as np
import sys, os
sys.path.append('../../shared/utilsGRF/')
from IPython.display import display
import writescripts 
```

```python
edges=[(1,'a1',2),(2,'a2',3),(3,'a3',1),
       (4,'epsilon1',5),(5,'epsilon2',6),(6,'epsilon3',4),
      (1,'kb1-x',4),(4,'ku1',1),(2,'kb2-x',5),(5,'ku2',2),(3,'kb3-x',6),(6,'ku3',3)]

MTTfolder='../../shared/utilsGRF' #folder with MTT.py 
basename='graph'

parlist=[x[1] for x in edges]
parlist=list(set(parlist))


list(set(parlist))

for pnum,par in enumerate(parlist):
    if '-x' in par:
        parlist[pnum]=parlist[pnum].replace('-x','')
print(parlist)

obj=writescripts.PrepareFilesNoneq(edgelist=edges,varGRF='x',concvars=['x'],parlist=parlist,MTTfolder=MTTfolder,graphbasename=basename)
obj.write_execute_parse()

obj.simpify_rhos()
```

```python
import sympy
from sympy.parsing.sympy_parser import parse_expr
```

```python
sympy.init_printing()
```

```python
for label in parlist:
    #print(label)
    exec("%s=sympy.symbols('%s')"%(label,label))
for var in ['x']:
    exec("%s=sympy.symbols('%s')"%(var,var))
a,alpha1,alpha2,alpha3,kb,ku=sympy.symbols('a,alpha_1,alpha_2,alpha_3,k_b,k_u')
all_rhos_sympy=[]
for rho in obj.all_rhos:
    expr=parse_expr(rho.split("=")[1])
    expr_simple=expr.subs({a1:a,a2:a,a3:a,epsilon1:alpha1*a,epsilon2:alpha2*a,epsilon3:alpha3*a,kb1:kb,kb2:kb,kb3:kb,ku1:ku,ku2:ku,ku3:ku})
    all_rhos_sympy.append(expr_simple)
```

```python
for i in range(6):
    print("Trees for node ",i+1)
    for tree in sympy.Add.make_args(all_rhos_sympy[i]):
        print(tree)
    print()
```

```python
B=sympy.symbols("B")
phi1=all_rhos_sympy[0].subs(kb*x,B)
phi2=all_rhos_sympy[1].subs(kb*x,B)
phi3=all_rhos_sympy[2].subs(kb*x,B)
phi4=all_rhos_sympy[3].subs(kb*x,B)
phi5=all_rhos_sympy[4].subs(kb*x,B)
phi6=all_rhos_sympy[5].subs(kb*x,B)
```

```python
phi1+phi2+phi3
```

```python
phi4+phi5+phi6
```

```python
num=phi3+alpha3*phi6
den=phi1+phi2+phi3+phi4+phi5+phi6
```

```python
dnum=sympy.diff(num,B)
dden=sympy.diff(den,B)
```

```python
polydnum=sympy.poly(dnum,B)
polynum=sympy.poly(num,B)
polyden=sympy.poly(den,B)
polydden=sympy.poly(dden,B)
```

```python
polyr=sympy.Poly(sympy.expand(dnum*den-num*dden),B)
polyr
```

```python
polynum.degree(),polydnum.degree(),polyden.degree(),polydden.degree()
```

```python
#coeff of degree 4:
sympy.expand(polynum.coeffs()[0]*polyden.coeffs()[1]-polyden.coeffs()[0]*polynum.coeffs()[1])
```

```python
#coeff of degree 3:
2*sympy.expand(polynum.coeffs()[0]*polyden.coeffs()[2]-polyden.coeffs()[0]*polynum.coeffs()[2])==polyr.coeffs()[1]
```

```python
An,Bn,Cn,Dn=polynum.coeffs()
Ad,Bd,Cd,Dd=polyden.coeffs()
```

```python
#coeff of degree 2:
sympy.expand(3*An*Dd+Bn*Cd-Cn*Bd-3*Dn*Ad)==polyr.coeffs()[2]
```

```python
#coeff of degree 1:
sympy.expand(2*(Bn*Dd-Dn*Bd))==polyr.coeffs()[3]
```

```python
#Degree 0
sympy.expand(Cn*Dd-Dn*Cd)==polyr.coeffs()[4]
```

```python
sympy.collect(sympy.expand(An*Dd-Ad*Dn),(a,ku)) #notice how each group of terms fulfills the condition
```

```python
sympy.collect(sympy.expand(Bn*Cd-Bd*Cn),(a,ku)) #notice how each group of terms fulfills the condition
```

```python
sympy.expand(Ad*Bn)
```

```python
I=a**2*alpha1*alpha2
K=a**2*ku*alpha1*alpha2
M=3*a**3*alpha1*alpha2
N=a**2+ku*(alpha1+alpha2)
L=a**2*ku*(alpha1*alpha2+alpha1*alpha3+alpha2*alpha3)
O=3*a**3*(alpha1*alpha2+alpha1*alpha3+alpha2*alpha3)
Q=a**2*ku*(alpha1+alpha2+alpha1+alpha3+alpha2+alpha3)
```

```python
sympy.expand(L+O+Q)==Bd
```

```python
sympy.expand(I*sympy.expand(O)-sympy.expand(M)*Ad)
```

```python
sympy.expand(alpha3*I*(L+Q)-Ad*(K+alpha3*N))
```
