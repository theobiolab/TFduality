```python
def nchoosek_sympy(m,k):
    return sympy.factorial(m)/(sympy.factorial(k)*sympy.factorial(m-k))
def c_formula_sympy2(n,k,r):
    f1=nchoosek_sympy(n-1,k)
    f2=nchoosek_sympy(k,r)
    num=n*f1*f2
    den=n-k+r
    return num/den 
```

```python
def nchoosek(n,k):
    #print("n",n,"k",k)
    return math.factorial(n)/(math.factorial(n-k)*math.factorial(k))
def c_formula(n,degree,nr):
    num=n*nchoosek(n-1,n-degree)*nchoosek(n-degree,nr)
    den=degree+nr
    return num/den
```

```python
n=20


for i in np.arange(1,n-2):
    for g in np.arange(2,n-i+1):
        if n-i-g<0:
            print("wrong g")
            raise ValueError
        k2=n-i-g
        k2_=k2-1
        k1_=n-i
        k1=k1_-1
        #sums <k2
        for S in np.arange(0,k2):
            left=0
            right=0
            for r2 in range(S+1):
                r1=S-r2
                f1=c_formula(n,n-k1,r1)
                f2=c_formula(n,n-k2,r2)
                f12=f1*f2
                left+=f12

                f1_=c_formula(n,n-k1_,r2)
                f2_=c_formula(n,n-k2_,r1)
                f12_=f1_*f2_
                right+=f12_
                #if f12<f12_:
                #    print("S:%d, r1:%d, r2:%d, k1:%d, k2: %d, k1_:%d, k2_:%d, f12:%d, f12_:%d"%(S,r1,r2,k1,k2,k1_,k2_,f12,f12_))
            if left-right<0:
                
                print("not fulfilled for i=%d, g=%d, S=%d"%(i,g,S))

    
```
