```r
library(reshape)
library(ggplot2)
```

```r
data<-read.csv("./expdata/MPRA/SuppData_withAf.csv")
data
```

```r
subdf<-data[c("Affinity.group","nrepeats","Mean")]
subdf1<-melt(subdf,id=c("nrepeats","Affinity.group"))
subdf1<-subdf1[order(subdf1$nrepeats),]
```

```r
subdf1$nrepeats
```

```r
myplot<-ggplot(subdf1, aes(x = nrepeats, y = value)) + facet_wrap(~as.factor(Affinity.group)) + geom_point() + geom_smooth( span=1) #+ scale_x_continuous("sum of affinities") + scale_y_continuous("activity") + theme_classic()
print(myplot)
```

```r
grouped_data <- split(subdf1, subdf1$Affinity.group)
```

```r
grouped_fits <- lapply(grouped_data, function(group) {
  loess_model <- loess(value ~ nrepeats, data = group, span = 1)
  group$loess_fit <- predict(loess_model)
  return(group)
})
fitted_data <- do.call(rbind, grouped_fits)
```

```r
unique_groups <- unique(fitted_data$Affinity.group)
colors <- c("red", "orange", "yellow","green","cyan","blue")

#plot(data$nrepeats, data$value, col = "gray", pch = 16, main = "LOESS Fits by Group", xlab = "X", ylab = "Y")

outputs<-list()
for (i in seq_along(unique_groups)) {
  group <- fitted_data[fitted_data$Affinity.group == unique_groups[i], ]
  plot(group$nrepeats, group$value, col = colors[i], pch = 16)
  lines(group$nrepeats, group$loess_fit, col = colors[i], lwd = 5)
  print(group$nrepeats)
  print(group$loess_fit)
  meang<-tapply(group$loess_fit, group$nrepeats, mean)
  print(meang)
    outputs[[i]]<-meang
}
```

```r
outputs
```

```r
as.data.frame(outputs[[1]],col.names=c("nrepeats","fit"))
```

```r
fit_output<-do.call(rbind, outputs)
fit_output <- cbind(fit_output, affinity= c(0.1,0.25,0.5,0.75,0.89,1))
```
